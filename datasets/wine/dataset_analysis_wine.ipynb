```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
# %pip install kagglehub
import kagglehub

# path = kagglehub.dataset_download("taweilo/wine-quality-dataset-balanced-classification")
# print("Path to dataset files:", path)

df = pd.read_csv('wine_data.csv') # https://www.kaggle.com/datasets/taweilo/wine-quality-dataset-balanced-classification
df = df.sample(frac=1).iloc[:2000, :].reset_index(drop=True)
df['quality'].value_counts()
```

```python
df['quality'] = df['quality'].apply(lambda x: 1 if x > 5 else 0)
df['quality'].value_counts()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1:]

print(df.shape)
print(X.shape) # no. of features is >= 50% of no. of observations
print(y.shape)
```

```python
display(X.head(3))
display(y.head(3))
```

```python

num_permutations = 99 # no. of observations is 2000, no. of features is 11 * 100  = 1100 (>= 50%)

perm_dfs = []

for i in range(1, num_permutations + 1):
    X_perm = X.apply(lambda col: np.random.permutation(col))
    X_perm.columns = [f'perm_{i}_{col}' for col in X.columns]
    perm_dfs.append(X_perm) 

X_final = pd.concat([X] + perm_dfs, axis=1)

display(X.head(3))
display(X_final.head(3))
```

```python
X = X_final
display(X.describe())
display(y.describe())
```

```python
X[X.isnull().any(axis=1)] # X does not contain any rows with null values
```

```python
threshold = 0.9
correlation_matrix = X.corr()

collinear_pairs = np.where((correlation_matrix > threshold) & (correlation_matrix < 1.0))

collinear_features = [(correlation_matrix.index[i], correlation_matrix.columns[j]) for i, j in zip(*collinear_pairs)]

print(collinear_features)
```

```python
to_drop = set()

for feature1, feature2 in collinear_features:
    to_drop.add(feature2) 

X_reduced = X.drop(columns=to_drop)
print(X.shape)
X_reduced.shape # no. of features is > 50% of no. of observations
```
