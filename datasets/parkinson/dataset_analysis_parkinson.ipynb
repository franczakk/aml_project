```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('pd_speech_features.csv') # https://archive.ics.uci.edu/dataset/470/parkinson+s+disease+classification
df.columns = df.iloc[0,:].tolist()
df = df.iloc[1:,1:]
for column in df.columns:
    df[column] = df[column].astype(float)

df = df.sample(frac=1, random_state=42).iloc[:600, :].reset_index(drop=True)

X = df.iloc[:,:-1]
y = df.iloc[:,-1:]

print(df.shape)
print(X.shape) # no. of features is >= 50% of no. of observations
print(y.shape)
```

```python
threshold = 0.9
correlation_matrix = X.corr()

collinear_pairs = np.where((correlation_matrix > threshold) & (correlation_matrix < 1.0))

collinear_features = [(correlation_matrix.index[i], correlation_matrix.columns[j]) for i, j in zip(*collinear_pairs)]

print(collinear_features)
```

```python
to_drop = set()

for feature1, feature2 in collinear_features:
    to_drop.add(feature2) 

X_reduced = X.drop(columns=to_drop)
print(X.shape)
X_reduced.shape # no. of features is > 50% of no. of observations
```
